```python
circle_1 = cv2.imread("0_1.jpg", cv2.IMREAD_GRAYSCALE)
circle_2 = cv2.imread("0_2.jpg", cv2.IMREAD_GRAYSCALE)
circle_3 = cv2.imread("0_3.jpg", cv2.IMREAD_GRAYSCALE)
circle_4 = cv2.imread("0_4.png", cv2.IMREAD_GRAYSCALE)
circle_5 = cv2.imread("0_5.jpg", cv2.IMREAD_GRAYSCALE)

Circle = [circle_1, circle_2, circle_3, circle_4, circle_5]

triangle_1 = cv2.imread("1_1.png", cv2.IMREAD_GRAYSCALE)
triangle_2 = cv2.imread("1_2.png", cv2.IMREAD_GRAYSCALE)
triangle_3 = cv2.imread("1_3.png", cv2.IMREAD_GRAYSCALE)
triangle_4 = cv2.imread("1_4.jpg", cv2.IMREAD_GRAYSCALE)
triangle_5 = cv2.imread("1_5.jpg", cv2.IMREAD_GRAYSCALE)

Triangle = [triangle_1, triangle_2, triangle_3, triangle_4, triangle_5]

square_1 = cv2.imread("2_1.jpg", cv2.IMREAD_GRAYSCALE)
square_2 = cv2.imread("2_2.jpg", cv2.IMREAD_GRAYSCALE)
square_3 = cv2.imread("2_3.png", cv2.IMREAD_GRAYSCALE)
square_4 = cv2.imread("2_4.png", cv2.IMREAD_GRAYSCALE)
square_5 = cv2.imread("2_5.png", cv2.IMREAD_GRAYSCALE)

Square = [square_1, square_2, square_3, square_4, square_5]

test = cv2.imread("test.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# define a threshold, 128 is the middle of black and white in grey scale
thresh = 200

# threshold the image
for i in range(5):
    Circle[i] = cv2.resize (Circle[i], (128, 128), interpolation = cv2.INTER_CUBIC)
    Circle[i] = cv2.threshold(Circle[i], thresh, 255, cv2.THRESH_BINARY)[1]
    Triangle[i] = cv2.resize (Triangle[i], (128, 128), interpolation = cv2.INTER_CUBIC)
    Triangle[i] = cv2.threshold(Triangle[i], thresh, 255, cv2.THRESH_BINARY)[1]
    Square[i] = cv2.resize (Square[i], (128, 128), interpolation = cv2.INTER_CUBIC)
    Square[i] = cv2.threshold(Square[i], thresh, 255, cv2.THRESH_BINARY)[1]

test = cv2.resize (test, (128, 128), interpolation = cv2.INTER_CUBIC)
test = cv2.threshold(test, thresh, 255, cv2.THRESH_BINARY)[1]
```

```python
test = Square[1]
```

```python
distance = []
for j in [Circle, Triangle, Square]:
    for i in j:
        buf = cv2.subtract(i, test)
        distance.append(np.sum(buf >= 200))
```

```python
for i in range(15):
    distance[i] = 10000000 / (1 + distance[i] ** 2)
```
